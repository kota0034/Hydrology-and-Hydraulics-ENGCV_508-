```python
import matplotlib.pyplot as plt

# Data
nodes = ['R', 'A', 'B', 'C', 'D', 'E', 'F']
dead_end_pressure = [196.2, 303.2099714, 249.7619524, 298.824945, 305.917903, 293.8867957, 307.4864732]
two_loop_pressure = [196.2, 251.6290651, 406.8404412, 238.2986295, 361.6142192, 375.6226593, 354.0058798]

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(nodes, dead_end_pressure, marker='o', label='Dead-End Pressure (kPa)', linestyle='--')
plt.plot(nodes, two_loop_pressure, marker='s', label='Two-Loop Pressure (kPa)', linestyle='-')

# Labels and Title
plt.xlabel("Node")
plt.ylabel("Pressure (kPa)")
plt.title("Pressure Comparison: Dead-End vs Two-Loop Hardy Cross")
plt.grid(True)
plt.legend()
plt.tight_layout()

# Show Plot
plt.show()
```
